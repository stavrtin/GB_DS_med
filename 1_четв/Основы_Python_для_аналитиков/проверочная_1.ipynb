```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.1. Скачать следующие данные: kc-house-data и laptop_price
df = pd.read_csv('kc-house-data.csv', sep=',')
# df = pd.read_csv('laptop_price.csv', sep=',', encoding='cp1251')
df
```

```python
df.columns
```

```python
#  1.2. Изучите стоимости недвижимости
df.agg({'price': ['sum', 'min', 'max']})
```

```python
# 1.2. Изучите стоимости недвижимости
plt.figure(figsize=(6, 4))
sns.histplot(df['price'], bins=600)
plt.title('Распределение стоимости квартир')
plt.xlabel('Стоимость, тыс.')
plt.ylabel('Кол-во квартир');
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df['price']);
```

```python
# 1.3. Изучите распределение квадратуры жилой
plt.figure(figsize=(6, 4))
sns.histplot(df['sqft_living'], bins=60)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Площадь жилая, м2.')
plt.ylabel('Кол-во квартир');
```

```python
# 1.4. Изучите распределение года постройки
df.groupby('yr_built').agg({'price':['max','mean'], 'sqft_living':['count']})
```

```python
# 1.4. Изучите распределение года постройки
plt.figure(figsize=(6, 4))
sns.histplot(df['yr_built'], bins=60)
plt.title('Распределение года постройки')
plt.xlabel('Площадь жилая, м2.')
plt.ylabel('Кол-во квартир')
plt.xticks(rotation=30);
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br>
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.<br>

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
# corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```
